```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = r"D:\fyp\DatasetGeneration\The smells\smells_react.csv"
df = pd.read_csv(file_path)

# Checking the Frequency of each present Smell
smell_counts = df["Smell"].value_counts()

# Create a custom color palette for the bars
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd']

# Plot the bar chart with custom styling
plt.figure(figsize=(12, 6))
ax = smell_counts.plot(kind="bar", color=colors, edgecolor='black', linewidth=1)

plt.xlabel("Smell")
plt.ylabel("Frequency")
plt.title("Frequency of Smells")

# Display grid lines
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Display the number of each smell
for i, count in enumerate(smell_counts):
    ax.text(i, count + 1, str(count), ha="center", va="bottom", fontsize=12, color="black")

plt.xticks(rotation=45, ha="right")
plt.tight_layout()

plt.show()
```

```python
import os

folder_path = "D:/fyp/DatasetGeneration/The smells/folders"  # Replace with the actual path to the folder

# Get the absolute path of the folder
folder_path = os.path.expanduser(folder_path)
folder_path = os.path.abspath(folder_path)

# Initialize counters and empty file list
total_files = 0
empty_files = 0
empty_file_list = []

# Iterate through the files in the folder
for file_name in os.listdir(folder_path):
    if file_name.endswith((".js",".tsx",".jsx")):
        file_path = os.path.join(folder_path, file_name)

        # Check if the file is empty
        if os.path.getsize(file_path) == 0:
            empty_files += 1
            empty_file_list.append(file_name)
        
        total_files += 1

# Print the results
print(f"Total number of files: {total_files}")
print(f"Total number of empty files: {empty_files}")

if empty_files > 0:
    print("Empty files:")
    for empty_file in empty_file_list:
        print(empty_file)
```
